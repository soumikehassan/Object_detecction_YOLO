```python
import cv2
import torch

# Load the pre-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # You can also use yolov5m, yolov5l, or yolov5x for larger models

# Path to the video file (you can upload a video file to Colab or use a sample video)
video_path = 'MOV_3876.mp4'  # Replace with your video file path

# Open the video stream
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Perform object detection on the current frame
    results = model(frame)  # This will detect objects in the frame

    # Render the results (draw bounding boxes on the frame)
    results.render()  # Draw the bounding boxes on the frame

    # Display the frame with the detection
    cv2_imshow(frame) # Use cv2_imshow instead of cv2.imshow
    
    # Break the loop if the 'q' key is pressed (for closing the window)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close any OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import torch

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # Choose a smaller model for real-time speed

# Open webcam (device 0 is the default webcam)
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform object detection
    results = model(frame)  # Pass the frame to the YOLO model for detection

    # Render the results (draw bounding boxes)
    results.render()  # Draw bounding boxes on the frame

    # Display the frame with bounding boxes
    cv2.imshow("Real-Time Object Detection", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
